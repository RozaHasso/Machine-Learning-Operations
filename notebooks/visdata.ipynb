```python
import torch
from torch.utils.data import DataLoader
import numpy as np
import os

import matplotlib.pyplot as plt
```

```python
print(os.getcwd())
train_set = torch.load("data/processed/train_dataset")
train_set = DataLoader(train_set, batch_size = 16)

examples = next(iter(train_set))

for label, img  in enumerate(examples):
   plt.imshow(img[1].squeeze(), cmap='Greys_r')
   plt.show()
   print(f"Label: {label}")
   break
```
